```python
import requests

def precision(TP, FP):
    """Precision = TP / (TP + FP)"""
    if TP + FP == 0:
        return 0
    return float(TP / (TP + FP))


def recall(TP, FN):
    """Recall = TP / (TP + FN)"""
    if TP + FN == 0:
        return 0
    return float(TP / (TP + FN))


def f1_score(precision, recall):
    """F1-Score = 2 * (precision * recall) / (precision + recall)"""
    if precision + recall == 0:
        return 0
    return float(2 * (precision * recall) / (precision + recall))


data1 = []

cases = [
    {
        "name": "title:Python",
        "tp": 1, "fp": 0, "fn": 0,
        "tp_langs": ["Python"],
        "fp_langs": [],
        "fn_langs": []
    },
    {
        "name": "title:Python3",
        "tp": 0, "fp": 0, "fn": 1,
        "tp_langs": [],
        "fp_langs": [],
        "fn_langs": ["Python"]
    },
    {
        "name": "title:Python3.12",
        "tp": 0, "fp": 0, "fn": 1,
        "tp_langs": [],
        "fp_langs": [],
        "fn_langs": ["Python"]
    },
    {
        "name": "title:Go",
        "tp": 1, "fp": 0, "fn": 0,
        "tp_langs": ["Go"],
        "fp_langs": [],
        "fn_langs": []
    },
    {
        "name": "title:Golang",
        "tp": 0, "fp": 0, "fn": 1,
        "tp_langs": [],
        "fp_langs": [],
        "fn_langs": ["Go"]
    },
    {
        "name": "document:functional",
        "tp": 13, "fp": 11, "fn": 0,
        "tp_langs": ["Elixir", "Haskell", "R", "Scala", "MATLAB", "Rust", "Zig", "Python", "JavaScript", "TypeScript", "C++", "Lua", "Go"],
        "fp_langs": ["Ruby", "Groovy", "Java", "Kotlin", "Swift", "Lua", "C#", "PHP", "Dart", "PowerShell", "C"],
        "fn_langs": []
    },
    {
        "name": "document:declarative",
        "tp": 10, "fp": 11, "fn": 3,
        "tp_langs": ["TypeScript", "Scala", "C++", "Go", "JavaScript", "Rust", "Lua", "Haskell", "Zig", "MATLAB"],
        "fp_langs": ["Kotlin", "Groovy", "C", "Swift", "PHP", "C#", "Ruby", "Assembly language", "Dart", "Java", "PowerShell"],
        "fn_langs": ["Elixir", "R", "Python"]
    },
    {
        "name": "document:object-oriented",
        "tp": 14, "fp": 5, "fn": 3,
        "tp_langs": ["Ruby", "Groovy", "Java", "Kotlin", "JavaScript", "Python", "Swift", "Go", "Lua", "C#", "PHP", "TypeScript", "Dart", "Rust"],
        "fp_langs": ["Scala", "C", "MATLAB", "R", "Assembly language"],
        "fn_langs": ["PowerShell", "Zig", "C++"]
    },
    {
        "name": "document:imperative",
        "tp": 5, "fp": 3, "fn": 12,
        "tp_langs": ["PowerShell", "Ruby", "Go", "C#", "Python"],
        "fp_langs": ["Elixir", "C", "Scala"],
        "fn_langs": ["Java", "PHP", "Dart", "Kotlin", "Groovy", "Swift", "Rust", "Zig", "JavaScript", "TypeScript", "C++", "Lua"]
    },
    {
        "name": "title:C#",
        "tp": 1, "fp": 2, "fn": 0,
        "tp_langs": ["C#"],
        "fp_langs": ["C", "C++"],
        "fn_langs": []
    },
    {
        "name": r'title:"C%20Sharp"',
        "tp": 1, "fp": 0, "fn": 0,
        "tp_langs": ["C#"],
        "fp_langs": [],
        "fn_langs": []
    },
    {
        "name": "title:C++",
        "tp": 2, "fp": 1, "fn": 0,
        "tp_langs": ["C++", "C"],
        "fp_langs": ["C#"],
        "fn_langs": []
    },
    {
        "name": "title:C plus plus",
        "tp": 0, "fp": 0, "fn": 2,
        "tp_langs": [],
        "fp_langs": [],
        "fn_langs": ["C", "C++"]
    },
    {
        "name": "document:\"JavaScript framework\"",
        "tp": 2, "fp": 1, "fn": 0,
        "tp_langs": ["JavaScript", "TypeScript"],
        "fp_langs": ["Dart"],
        "fn_langs": []
    },
    {
        "name": "document:\"Javascrit framework\"",
        "tp": 0, "fp": 0, "fn": 2,
        "tp_langs": [],
        "fp_langs": [],
        "fn_langs": ["JavaScript", "TypeScript"]
    },
    {
        "name": "document:\"most modern programming languages\"",
        "tp": 0, "fp": 0, "fn": 8,
        "tp_langs": [],
        "fp_langs": [],
        "fn_langs": ["Zig", "Rust", "Kotlin", "TypeScript", "Elixir", "Dart", "Swift", "Go"]
    },
    {
        "name": "document:\"modern programming languages\"",
        "tp": 1, "fp": 0, "fn": 7,
        "tp_langs": ["Swift"],
        "fp_langs": [],
        "fn_langs": ["Zig", "Rust", "Kotlin", "TypeScript", "Elixir", "Dart", "Go"]
    },
    {
        "name": "document:\"trendy programming languages\"",
        "tp": 0, "fp": 0, "fn": 8,
        "tp_langs": [],
        "fp_langs": [],
        "fn_langs": ["Zig", "Rust", "Kotlin", "TypeScript", "Elixir", "Dart", "Swift", "Go"]
    },
    {
        "name": "document:\"What is the best programming language for web development\"",
        "tp": 0, "fp": 0, "fn": 2,
        "tp_langs": [],
        "fp_langs": [],
        "fn_langs": ["JavaScript", "TypeScript"]
    },
    {
        "name": "document:\"Top programming languages for web development\"",
        "tp": 0, "fp": 0, "fn": 2,
        "tp_langs": [],
        "fp_langs": [],
        "fn_langs": ["JavaScript", "TypeScript"]
    },
    {
        "name": "document:\"Best coding languages for building websites\"",
        "tp": 0, "fp": 0, "fn": 2,
        "tp_langs": [],
        "fp_langs": [],
        "fn_langs": ["JavaScript", "TypeScript"]
    },
    {
        "name": "document:\"Best programming language for web development\"",
        "tp": 0, "fp": 0, "fn": 2,
        "tp_langs": [],
        "fp_langs": [],
        "fn_langs": ["JavaScript", "TypeScript"]
    },
    {
        "name": "document:\"Best systems programming language for safety\"",
        "tp": 0, "fp": 0, "fn": 4,
        "tp_langs": [],
        "fp_langs": [],
        "fn_langs": ["Rust", "Zig", "Swift", "Go"]
    },
    {
        "name": "document:\"What is a memory-safe programming language\"",
        "tp": 0, "fp": 0, "fn": 3,
        "tp_langs": [],
        "fp_langs": [],
        "fn_langs": ["Rust", "Zig", "Swift"]
    },
    {
        "name": "document:\"Most secure programming language\"",
        "tp": 0, "fp": 0, "fn": 1,
        "tp_langs": [],
        "fp_langs": [],
        "fn_langs": ["Rust"]
    },
]

for case in cases:
    s1_pre = precision(case["tp"], case["fp"])
    s1_rec = recall(case["tp"], case["fn"])
    s1_f1 = f1_score(s1_pre, s1_rec)
    case_out = {
        "name": case["name"],
        "s1_pre": s1_pre,
        "s1_rec": s1_rec,
        "s1_f1": s1_f1,
        "tp": case["tp"],
        "fp": case["fp"],
        "fn": case["fn"],
        "tp_langs": case["tp_langs"],
        "fp_langs": case["fp_langs"],
        "fn_langs": case["fn_langs"],
    }
    data1.append(case_out)
    print(f"[{case['name']}] F1-Score {s1_f1:.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- Adapter: Gruppiere results nach core ---
from collections import defaultdict

grouped = defaultdict(list)
for r in results:
    grouped[r["core"]].append(r)

# Gehe davon aus, dass genau zwei Cores da sind:
core_names = list(grouped.keys())
if len(core_names) != 2:
    raise ValueError("Erwarte genau 2 Cores, gefunden: " + ", ".join(core_names))

core1, core2 = core_names[0], core_names[1]
data1 = grouped[core1]
data2 = grouped[core2]

# Convert lists to dicts keyed by name
d1 = {d["name"]: d for d in data1}
d2 = {d["name"]: d for d in data2}

# Metrics and their labels
metrics = ["s1_pre", "s1_rec", "s1_f1"]
labels = ["Precision", "Recall", "F1 Score"]

# Plot improvement for each name
for name in d1:
    # raw scores + diff
    v1 = [d1[name][m] for m in metrics]
    v2 = [d2[name][m] for m in metrics]
    diff = [b - a for a, b in zip(v1, v2)]

    # build the plot
    x = np.arange(len(metrics))
    width = 0.25
    fig, ax = plt.subplots(figsize=(12, 4))

    ax.bar(x - width,   v1,   width, label="Data 1")
    ax.bar(x,           v2,   width, label="Data 2")
    ax.bar(x + width,   diff, width, label="∆ (2-1)")

    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.set_ylim(
        min(min(diff), 0) - 0.05,
        max(v1 + v2 + diff) + 0.05
    )
    ax.set_ylabel("Score / Improvement")
    ax.set_title(f"Scores & Improvement for {name}")
    ax.legend(loc="upper left")

    # pull out TP/FP/FN from dataset 1 (or 2) and make the little comment box
    tp1, fp1, fn1 = d1[name]["tp"], d1[name]["fp"], d1[name]["fn"]
    tp2, fp2, fn2 = d2[name]["tp"], d2[name]["fp"], d2[name]["fn"]
    comment = (
        "Data 1\n"
        f"TP: {tp1}\n"
        f"FP: {fp1}\n"
        f"FN: {fn1}\n"
        "Data 2\n"
        f"TP: {tp2}\n"
        f"FP: {fp2}\n"
        f"FN: {fn2}\n"
        "TP = Richtige Dokumente gefunden\n"
        "FP = Gefundene, aber falsche Dokumente\n"
        "FN = Nicht gefundene, aber relevante Dokumente\n"
    )
    # place text to the right of the axes
    ax.text(1.02, 0.5, comment,
            transform=ax.transAxes,
            va="center", ha="left", fontsize=8,
            bbox=dict(boxstyle="round,pad=0.3", facecolor="white", edgecolor="#666"))

    plt.tight_layout()
    plt.show()
```
